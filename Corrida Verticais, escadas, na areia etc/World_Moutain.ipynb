```python
import pandas as pd

# Suppose you have a DataFrame called 'df' with columns 'Name' and 'Team'
df = pd.read_csv(url2, encoding='latin-1')  # Use this line to read from a CSV file

df = df[df['GÉNERO'] == 'M']

# Define the regex pattern
pattern = r'Sénior|M\d+'

# Filter the rows where the 'Escalão' column matches the pattern
df = df[df['ESCALÃO / CATEGORY'].str.contains(pattern)]

# Remove the team with the name 'Individual'
df_filtered = df[df['CLUBE / TEAM'] != 'Individual']

# Group the filtered data by 'Team' and get the number of athletes in each team
grouped = df_filtered.groupby('CLUBE / TEAM').size().reset_index(name='Number of Athletes')

# Sort the grouped DataFrame by the number of athletes in descending order
sorted_grouped = grouped.sort_values(by='Number of Athletes', ascending=False)

# Iterate over the teams with more than three athletes and display the names of athletes
for team, num_athletes in sorted_grouped[sorted_grouped['Number of Athletes'] >= 3].itertuples(index=False):
    team_athletes = df_filtered[df_filtered['CLUBE / TEAM'] == team]['NOME / NAME']
    print(f"Team: {team}")
    print("Athletes:")
    for athlete in team_athletes:
        print(athlete)
    print()
```
